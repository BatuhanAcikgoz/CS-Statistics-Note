## 1. Temel Girdi, Çıktı ve Aritmetik İşlemler

`clc` komut penceresini temizler, `clear` çalışma alanındaki (workspace) değişkenleri siler, `close` açık olan tüm grafikleri kapatır.

`fprintf` fonksiyonu, formatlanmış metin ve değişkenleri ekrana basmak için kullanılır. `%s` metin (string), `%d` tamsayı (integer), `%f` ondalıklı sayı (float) için yer tutucudur.

```octave
clc
clear
close
isim = "ali";
fprintf('abimin adı %s dir.\n', isim)


x = input('x değerini giriniz: \n');
y = input('y değerini giriniz: \n');
fprintf('x ve y nin toplamı = %d \n', x + y)
fprintf('x ve y nin çarpımı = %d \n', x * y)
fprintf('x ve y nin bölümü = %d \n', x / y)

% Aşağıdaki satır, tek bir fprintf içinde birden fazla sonuç basmayı gösterir
%fprintf('x ve y nin toplamı = %d \n x ve y nin çarpımı = %d \n',x+y,x*y)
```

## 2. `for` Döngüsü Örneği (Zar Atma)

Bu örnek, 6 gelene kadar zar atma simülasyonu yapar.

Bir `for` döngüsü (`1:n`) belirlenen deneme sayısına kadar çalışır. `unidrnd(6)` 1'den 6'ya kadar rastgele bir tamsayı (zar) üretir.

Eğer atış 6 gelirse, `break` komutu döngüyü (deneme sayısı `n`'e ulaşmamış olsa bile) erken sonlandırır.

```octave
clc; clear; close;

n = input('yeterince büyük deneme sayısı giriniz: \n');

for i = 1:n
    x(i) = unidrnd(6); % 1-6 arası rastgele tamsayı
    fprintf('%d. atış: %d \n', i, x(i));
    
    if x(i) == 6 
        break % Döngüyü kır
    end
end

fprintf('ilk kez 6, %d. atışta geldi\n', i);
```

## 3. `while` Döngüsü Örneği (Zar Atma)

Bu örnek, `for` döngüsü ile yapılan bir önceki işlemin aynısını `while` döngüsü ile yapar.

`while x ~= 6` koşulu, "x değişkeni 6'ya eşit olmadığı (`~=`) sürece" döngüye devam et anlamına gelir. Bir sayaç (`say`) her döngüde manuel olarak artırılır.

```octave
clc; clear; close;

say = 0; % Sayaç
x = 1;   % Döngünün başlayabilmesi için x'e 6 olmayan bir ilk değer atanır

while x ~= 6
    x = unidrnd(6);
    say = say + 1;
    fprintf('%d. atış: %d \n', say, x);
end

fprintf('ilk kez 6, %d. atışta geldi\n', say);
```

## 4. `fprintf` ile String ve Sayısal Değerleri Birleştirme

Bu blok, bir sayının karesini ve küpünü hesaplar, `string()` fonksiyonu ile oluşturulan metinlerle birlikte `fprintf` kullanarak ekrana basar.

`\t` karakteri, çıktıda bir 'Tab' (sekme) boşluğu bırakır.

```octave
clc; clear; close;

x = input('bir değer giriniz:');
a = string('kare');
b = string('küp');

fprintf('%d %s: %d, %d %s: %d \n', x, a, x^2, x, b, x^3);

% \t (tab) kullanarak daha düzenli bir çıktı:
fprintf('%d %s: %d,\t %d %s: %d \n', x, a, x^2, x, b, x^3);
```

## 5. `disp` ile Çıktı Verme (Alternatif Yöntem)

`disp()` fonksiyonu, `fprintf`'e alternatiftir ancak sadece tek bir (genellikle metin/char) argüman alır.

Sayısal değerleri (`x+y`) metinle birleştirmek için `num2str()` (Number to String) fonksiyonu kullanılır. `[...]` köşeli parantezleri, bu metin parçalarını birleştirip tek bir metin (char array) oluşturur.

```octave
clc; clear; close;

x = input('x değerini giriniz: \n');
y = input('y değerini giriniz: \n');

disp(['x+y işleminin sonucu ' num2str(x + y) ' dir'])
```

## 6. Proje: Vektör İstatistikleri Hesaplama

Girilen bir vektör için ortalama, varyans, en büyük ve en küçük değerini bulup ekrana yazdıran bir program.

### 6a. Vektör Girdisinin Alınması

Önce vektörün boyutu (`n`) istenir, ardından `for` döngüsü ile `n` adet eleman tek tek kullanıcıdan alınır ve `x` vektörüne atanır.

```octave
clc; clear; close;

n = input('vektörün boyutunu giriniz:');
for i = 1:n
    disp([num2str(i), '. eleman: '])
    x(i) = input('');
end
```

### 6b. Vektörün Ekrana Yazdırılması

Girilen vektörün elemanları kontrol amaçlı ekrana yazdırılır. `...` (üç nokta), MATLAB'de kodun bir alt satırdan devam ettiğini gösteren satır devamı (line continuation) işaretidir.

```octave
for i = 1:n
    disp(['vektörün ', num2str(i), '. elemanı ', ...
        num2str(x(i)), ' dir']);
end
```

### 6c. İstatistiklerin Manuel Hesaplanması

MATLAB'in `mean()`, `var()`, `min()`, `max()` gibi hazır fonksiyonları olsa da, bu blokta bu istatistiklerin `for` döngüleri kullanılarak manuel olarak nasıl hesaplandığı gösterilmektedir.

1.  **Ortalama (Mean):** Elemanların toplamı (`top`) bulunur ve eleman sayısına (`n`) bölünür.
2.  **Varyans (Variance):** Her elemanın ortalamadan farkının karesi toplanır (`top2`) ve (örneklem varyansı için) `n-1`'e bölünür.
3.  **En Küçük (Min):** İlk eleman en küçük (`enk`) kabul edilir ve döngüde daha küçüğü aranır.
4.  **En Büyük (Max):** İlk eleman en büyük (`enb`) kabul edilir ve döngüde daha büyüğü aranır.

```octave
% Ortalama Hesabı
top = 0;
for i = 1:n
    top = top + x(i);
end
ort = top / n;

% Varyans Hesabı (Örneklem Varyansı)
top2 = 0;
for i = 1:n
    top2 = top2 + (x(i) - ort)^2;
end
vary = top2 / (n - 1);

% En Küçük Değer (min) Hesabı
enk = x(1);
for i = 2:n
    if x(i) < enk
        enk = x(i);
    end
end

% En Büyük Değer (max) Hesabı
enb = x(1);
for i = 2:n
    if x(i) > enb
        enb = x(i);
    end
end
```

### 6d. Sonuçların Yazdırılması

Hesaplanan tüm istatistiksel değerler `fprintf` ile ekrana basılır. Ondalıklı çıktılar için `%f` formatlayıcısı kullanılmıştır.

```octave
fprintf('Ortalama: %f \n', ort);
fprintf('Varyans: %f \n', vary);
fprintf('En Küçük: %f \n', enk);
fprintf('En Büyük: %f \n', enb);
```
