### RStudio Arayüzü Hakkında Notlar

R kodları genellikle RStudio gibi bir IDE (Geliştirme Ortamı) üzerinden yazılır. Arayüz genelde 4 ana bölümden oluşur:

1.  **Kaynak (Source) Alanı:** `.R` script (betik) dosyalarının açıldığı ve kodların yazıldığı yer (Sol üst).
2.  **Konsol (Console) Alanı:** Kodların çalıştırıldığı ve anlık çıktıların görüldüğü yer (Sol alt).
3.  **Çevre (Environment) Alanı:** Oluşturulan nesnelerin (değişkenler, veri setleri vb.) listelendiği yer (Sağ üst).
4.  **Dosyalar/Grafikler/Paketler Alanı:** Bilgisayardaki dosyaları görme, grafikleri (plots) görüntüleme, paketleri (packages) yükleme ve yardım (help) sekmelerinin bulunduğu yer (Sağ alt).

## Veri Turleri

Bir nesnenin veri turunun ne oldugunu `class()` fonksiyonu ile ogreniriz.

### Numeric (Sayısal - Kesirli)

Kesirli veriler (ondalıklı sayılar) R'da 'numeric' olarak adlandırılır.

```r
k = 24.3

class(k)
```

#### Veri Tipi Dönüşümü: `as.numeric()`

Veriyi 'numeric' hale donusturmek icin `as.numeric()` fonksiyonu kullanilir. Özellikle tırnak içinde (" ") yazılan ve aslında sayı olan değerler (yani 'character' - karakter) için bu gereklidir.

```r
t <- 45.6
class(t)
```

```r
u <- "45.6" # Tırnak içinde olduğu için bu bir karakter dizisidir (metin)
class(u)
```

'u' değişkenini sayısal ('numeric') hale dönüştürmemiz gerekiyor:

```r
u <- as.numeric(u)
class(u)
```

#### Veri Tipi Kontrolü: `is.numeric()`

Bir nesnenin 'numeric' olup olmadığını sorgulamak için `is.numeric()` fonksiyonu kullanılır. Sonuç `TRUE` (Evet) veya `FALSE` (Hayır) olarak döner.

```r
is.numeric(t) # TRUE evet, #FALSE hayir demek
```

### Integer (Tamsayı)

R, tamsayıları da ('numeric' gibi) sayısal olarak görse de, özel olarak 'integer' (tamsayı) olarak da tanımlayabiliriz.

```r
s <- 50
class(s) # R, bunu otomatik olarak 'numeric' olarak atar
```

```r
# Tamsayı (integer) olduğunu belirtmek için:
s <- as.integer(s)
class(s)
```

Kesirli bir sayiyi (numeric) tam sayiya donusturursek (integer), ondalık kısım atılır (yuvarlama yapılmaz, kesilir):

```r
y <- 24.3
class(y)
```

```r
y <- as.integer(y)
y
```

```r
class(y)
```

Tırnak içindeki bir tamsayıyı da dönüştürebiliriz:

```r
yas <- "36"
class(yas)
```

```r
yas <- as.integer(yas)
yas
```

Sayısal olmayan (metin) bir degeri tamsayı yapmaya kalkarsak `NA` (Not Available - Kayıp Veri) uyarısı alırız:

```r
yas <- "otuz alti"
yas <- as.integer(yas)
```

`TRUE` ve `FALSE` mantıksal değerleri tamsayıya dönüştürülebilir. `TRUE` 1, `FALSE` 0 olur.

```r
durum <- TRUE
as.integer(durum)
```

```r
durum <- FALSE
as.integer(durum)
```

#### Veri Tipi Kontrolü: `is.integer()`

Tamsayi olup olmadigini `is.integer()` fonksiyonu ile sorgulayabiliriz.

```r
x <- as.integer(200)
is.integer(x)
```

```r
class(x)
```

`typeof()` fonksiyonu ile de veri turu (daha teknik detayda) dogrudan ogrenilebilir:

```r
a <- as.integer(25)
typeof(a)
```

### Character (Karakter Dizisi / Metin)

Tırnak işaretleri (`"` veya `'`) içinde yazılan her şey karakter dizisidir.

```r
mesaj <- "merhaba gencler"
class(mesaj)
```

#### Veri Tipi Dönüşümü: `as.character()`

Sayısal bir degeri karaktere (metne) cevirmek icin `as.character()` kullanılır.

```r
n <- 125
class(n)
```

#### Veri Tipi Kontrolü: `is.character()`

`is.character()` karakter biciminde olup olmadigini kontrol etmek icin kullanılır.

### Logical (Mantiksal Veri)

Mantiksal degiskenler sadece iki deger icerebilir: `TRUE` (Doğru) ve `FALSE` (Yanlış).
Bunlar tırnaksız yazılmalıdır. `T` ve `F` kısaltmaları da kullanılabilir.

```r
b <- TRUE
class(b)
```

#### Veri Tipi Dönüşümü: `as.logical()`

Sayısal değişkenleri mantıksal veri tipine dönüştürebiliriz.
R'da `0` (sıfır) `FALSE` olarak, `0` dışındaki *tüm sayılar* (pozitif veya negatif) `TRUE` olarak kabul edilir.

```r
mesaj <- 1
as.logical(mesaj)
```

```r
mesaj <- 2
as.logical(mesaj)
```

```r
# 0 degerini icereni de mantiksal degere donusturelim
mesaj <- 0
as.logical(mesaj)
```

Tırnak içinde "TRUE" veya "FALSE" yazan karakter dizileri de mantıksal değere dönüştürülebilir.

```r
mesaj <- "TRUE"
as.logical(mesaj)
```

```r
class(mesaj) # Orijinal 'mesaj' hala karakter, dönüşüm yeni bir değer yaratır
```

#### Veri Tipi Kontrolü: `is.logical()`

`is.logical()` mantiksal veri turune sahip olup olmadigini gormek icin kullanılır.

```r
sonuc <- T # 'T' harfi 'TRUE' için kısaltmadır
is.logical(sonuc)
```

## Mantiksal Islemler (Operatörler)

Bu operatörler her zaman `TRUE` veya `FALSE` sonucu döndürür.

* `<` : Kucuktur
* `>` : Buyuktur
* `<=` : Esit veya kucuktur
* `>=` : Esit veya buyuktur
* `==` : Esittir (Atama operatörü `=` veya `<-` ile karıştırılmamalıdır!)
* `!=` : Esit degildir
* `||` : Veya
* `&` : Ve
* `!` : Degil (Bir ifadenin tersini alır)

Şimdi bu operatörleri deneyelim:

```r
k <- 125
f <- 220
```

```r
# k, f'den büyük mü?
k > f
```

```r
# k, f'ye eşit mi?
k == f
```

```r
# k, f'ye eşit değil mi?
k != f
```

```r
# k, f'den büyük mü? (Tekrar)
k > f
```
