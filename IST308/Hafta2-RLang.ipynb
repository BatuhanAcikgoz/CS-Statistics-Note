## Vektörler (Vectors)

Vektörler, R'daki en temel veri yapısıdır. Mutlaka bir ismi olmalıdır (bir değişkene atanmalıdır).

Vektör oluşturmak için `c()` (combine/concatenate) fonksiyonu kullanılır.

````r
a <- c(1, 2, 3, 4, 5)
````

Vektör oluştu. İçeriğine `show` ile (veya sadece adını yazarak) bakabiliriz:

````r
show(a)
````

Bir vektöre değer atamak için `assign()` fonskiyonu da kullanılabilir. Bu, `<-` operatörüne bir alternatiftir (genellikle programatik olarak değişken adı oluştururken kullanılır).

İlk argüman değişken adını *tırnak içinde* alır.

````r
assign("a", c(1, 2, 3, 4, 5))
assign("i", c(3, 5, 7, 9, 11))
````

Bir vektör oluşturulurken, `c()` fonksiyonu içinde başka bir vektör de kullanılabilir. Elemanlar birleştirilir.

````r
b <- c(6, 7, a)
show(b)
````

#### Vektörlerde Veri Tipi Zorlaması (Coercion)

Bir vektör sayısal, karakter veya mantıksal değer içerebilir.

**ANCAK:** Bir vektörde **sadece bir türden** veri yer alabilir.

Eğer farklı türlerde veri (örn: sayı ve metin) aynı vektöre kaydedilmek istenirse, R bu değerleri ortak bir paydaya (genellikle en esnek olan 'character' tipine) zorlar (type coercion).

````r
s <- c(25, 12.3, FALSE, "burak")
s # Görüldüğü üzere, R sayıları (25, 12.3) ve mantıksal değeri (FALSE) da tırnak içine aldı.
````

Vektörün sınıfını (class) kontrol edelim:

````r
class(s) # 'character' (karakter/metin) oldu.
````

## Vektör İndeksleme (Indexing)

Vektörler bir boyutlu bir dizidir ve bir indekse sahiptir. İndeks, vektör içindeki her bir değerin hangi pozisyonda olduğunu belirtir.

**ÖNEMLİ:** R'da indeksler (diğer birçok dilin aksine) **1'den başlar** (0'dan değil).

Bu indeks yardımıyla doğrudan vektör elemanlarına `[]` (köşeli parantez) kullanarak ulaşmak mümkündür.

````r
v <- c(5, 7, 3, 1, 0, -3)
v
````

`v` vektörünün 4. konumdaki elemanını görüntüleyelim:

````r
v[4] # 4. değer 1 olduğu için o gelir
````

Negatif indeksleme, o elemanı *hariç tutmak* (silmek) anlamına gelir. 

`v[-1]`, birinci sıradaki değeri (5) *göstermez* (çıktıdan çıkarır):

````r
v[-1]
````

Bu silme işlemi geçicidir. Değişikliğin kalıcı olması için, sonucun tekrar `v` değişkenine atanması gerekir:

````r
v <- v[-1] # v'nin yeni değeri, v'nin 1. eleman hariç halidir
v
````

**Not:** Aşağıdaki örneklerin orijinal script'teki yorumlarla tutarlı olması için `v` vektörünü orijinal haline döndürelim.

````r
v <- c(5, 7, 3, 1, 0, -3)
v
````

Vektörün boyutu dışında kalan bir indeksi çağırmayı deneyelim:

````r
v[8] # Olmadigi icin NA (Not Available/Mevcut Değil) gelecektir.
````

Birden çok (ama sırasız) değeri görüntülemek istersek, indeksleri bir vektör `c()` içinde veririz:

````r
v[c(1, 5)] # 1. ve 5. değeri getir (5 ve 0)
````

Birden çok elemanı *sıralı* bir şekilde (bir aralık olarak) çağırmak için `:` operatörü kullanabiliriz:

````r
v[c(1:4)] # 1. değerden 4. değere kadar (1, 2, 3, 4) olanları getirir.

# Not: v[1:4] şeklinde 'c()' olmadan yazmak daha yaygın bir kısayoldur.
````

#### Vektör Elemanlarını Değiştirme (Atama)

Bir vektörün içini (veya belirli indekslerini) başka bir değerle doldurabiliriz. 

Eğer atanan değer (sağ taraf) sol taraftaki indeks sayısından azsa, R 'recycling' (geri dönüşüm/tekrarlama) yapar. (Örn: 10 elemanı 1 elemanla değiştirmek isterseniz, o 1 elemanı 10 kez tekrarlar).

Burada `v` vektörünün (şu an 6 elemanlı) 1'den 10'a kadar olan indekslerine `2` atamaya çalışıyoruz.

R, 1'den 6'ya kadar olan elemanları `2` ile değiştirir ve mevcut olmayan 7, 8, 9, 10. elemanları `2` olarak *yeni oluşturur*.

````r
v[c(1:10)] <- 2 # 10 elemanın hepsi 2 olacaktır
v
````

Şimdi de bu (artık 'numeric' olan) vektörün 1-10 arası elemanlarını 'nurullah' metni ile değiştirelim.

Veri tipi 'character' olmaya zorlanacaktır (coercion).

````r
v[c(1:10)] <- "nurullah"
v
````

## Vektör Aritmetiği

Aritmetik işlemler (çarpma, toplama vb.) vektörler üzerinde eleman eleman (element-wise) çalışır.

#### Vektörün bir sayı ile çarpımı (Scalar multiplication):

Vektörün *her bir elemanı* o sayı ile çarpılır.

````r
d <- c(3, 5, 7, 9, -3)
d * 5
````

#### İki vektörün toplanması:

Eşleşen indekslerdeki elemanlar toplanır (1. ile 1., 2. ile 2., vb.).

**Not:** Vektörlerin uzunlukları aynı olmalıdır. Değilse, R 'recycling' yapar (kısa olanı tekrarlar) ve bu genellikle istenmeyen bir duruma ve bir uyarıya (warning) yol açar.

````r
# Hatırlatma: a = c(1, 2, 3, 4, 5)
# d = c(3, 5, 7, 9, -3)
z <- d + a
z # Çıktı: (3+1), (5+2), (7+3), (9+4), (-3+5)
````

#### İki vektörün çarpımı (Element-wise):

````r
t <- d * a
t # Çıktı: (3*1), (5*2), (7*3), (9*4), (-3*5)
````

## Mantıksal Vektörler (Logical Vectors)

Mantıksal değerler `TRUE` (veya `T`), `FALSE` (veya `F`) ve `NA` (kayıp veri) olabilir.

````r
mv <- c(T, F, T, NA)
````

Bir koşula bağlı olarak (karşılaştırma operatörleri kullanarak) mantıksal bir vektör elde edebiliriz.

Örn: `k` vektöründeki elemanlardan 3'ten büyük olanları `TRUE`, diğerlerini `FALSE` yapalım.

````r
k <- c(5, 7, 3, 1, 0, -1)
k > 3
````

#### Mantıksal Operatörler

Birden fazla koşulu birleştirmek için mantıksal operatörler kullanılır:

* `&` : VE (Element-wise AND)
* `|` : VEYA (Element-wise OR) 
* `!` : DEĞİL (NOT)
* `==` : Eşittir
* `!=` : Eşit değildir

(Not: `||` ve `&&` operatörleri de vardır ancak bunlar vektörün sadece *ilk* elemanını kontrol eder, genellikle `if` koşullarında kullanılır. Vektörel filtreleme işlemleri için `|` ve `&` kullanılır.)

Koşul: `k`'nın 3'ten küçük VE (`&`) 0'a eşit olmayan elemanları:

````r
k < 3 & k != 0
````

Koşul: `k`'nın 3'ten küçük VEYA (`|`) 0'a eşit olmayan elemanları:

````r
k < 3 | k != 0
````

## Vektör Üyelerini İsimlendirme (Named Vectors)

R'da vektörlerin elemanlarına 'names' (isimler) atanabilir. Bu, indekse (1, 2, 3...) ek olarak metin etiketleriyle de elemanlara erişmemizi sağlar.

R, vektör uzunluğu (10) ile isim vektörü uzunluğu (5) eşleşmediği için bir uyarı (warning) verecek ve isim vektörünü (5 elemanlı) 10 elemana tamamlamak için geri dönüştürecek (recycling) -> `c("ID1", ..., "ID99", "ID1", ..., "ID99")`

## Vektörlerde Sorgulama (Subsetting)

Köşeli parantez `[]` içine sayısal indeksler (1, 5, vb.) yerine, mantıksal bir vektör (TRUE/FALSE) vererek filtreleme yapabiliriz. Buna 'logical subsetting' denir.

````r
v <- c(5, 7, 3, 1, 0, -2)
````

`v[v > 0]` işlemi R'da şöyle çalışır:

1.  `v > 0` koşulu çalışır ve `c(TRUE, TRUE, TRUE, TRUE, FALSE, FALSE)` mantıksal vektörünü üretir.
2.  Bu vektör `v[]` içine verilir: `v[c(TRUE, TRUE, TRUE, TRUE, FALSE, FALSE)]`
3.  R, sadece `TRUE` olan pozisyonlardaki (1, 2, 3, 4) değerleri döndürür.

````r
v[v > 0] # 0'dan büyük olanlar
````

````r
v[v >= 0] # 0'a eşit veya 0'dan büyük olanlar
````

Aynı sorgulama karakter (metin) vektörleri için de geçerlidir:
(İsim vektörünü (isimsiz haline) yeniden oluşturalım)

````r
isim <- c("nur", "efe", "nuri", "nazlican", "celebi", "yusuf", "fevzullah", "talip", "emre", "melda")
isim[isim == "nur"]
````

## Vektör İçinden Değer Silmek

Daha önce gördüğümüz gibi, negatif indeksleme o elemanı *hariç tutar*.

````r
v <- c(5, 7, 6, 1, 0, -2)
v[-3] # 3. değeri (6) sil (geçici olarak, sadece çıktıda)
````

Görünürde `v` vektöründen 3. değer silindi. Vektörü tekrar çağıralım:

````r
v # Orijinal vektör değişmedi.
````

Değişikliğin kalıcı olabilmesi için elde edilen sonucun (çıktının) tekrar `v` vektörüne atanması (üzerine yazılması) gerekir.

````r
v <- c(5, 7, 6, 1, 0, -2)
v <- v[-c(1:4)] # 1'den 4'e kadar olan değerleri (5, 7, 6, 1) sil
v
````

#### Koşullu Güncelleme

Güncellemeler toplu olarak (vektörel) yapılabilir. Bunun için vektör indeksi (sorgulama bölümünde gördüğümüz gibi) bir koşul ifadesi olmalıdır.

````r
v <- c(5, 7, 6, 1, 0, -2)
v[v < 3] <- 0 # v'nin 3'ten küçük olan elemanlarını (1, 0, -2) 0'a dönüştür
v
````

## İstatistiksel Vektör Fonksiyonları

## Vektörün Sıralanması

#### `sort()`: Değerleri Sıralama

````r
v <- c(5, 7, 6, 1, 0, -2)
sort(v) # Varsayılan olarak küçükten büyüğe sıralar
````

Büyükten küçüğe sıralama yapmak için `decreasing = TRUE` argümanı kullanılır:

````r
v <- c(5, 7, 6, 1, 0, -2)
sort(v, decreasing = TRUE)
````

#### `order()`: İndeksleri Sıralama

`order()` fonksiyonu `sort()`'tan farklı çalışır: Değerleri değil, o değerlere ulaşmak için gereken *indekslerin sırasını* döndürür.

Yani, "En küçük eleman kaçıncı indekste? Ondan sonraki en küçük eleman kaçıncı indekste?" sorularının cevabını verir.

````r
v <- c(2, 8, 12, 6, 1, 13)
ind <- order(v)
ind

# Çıktı Yorumlaması: [5, 1, 4, 2, 3, 6]
# 5. eleman (1) en küçük
# 1. eleman (2) ikinci en küçük
# 4. eleman (6) üçüncü en küçük
# 2. eleman (8) dördüncü en küçük
# 3. eleman (12) beşinci en küçük
# 6. eleman (13) altıncı en küçük
````

Bu `ind` (indeks) vektörünü `v` üzerinde kullanırsak, `v`'nin sıralanmış halini elde ederiz:

````r
v[ind] # v[c(5, 1, 4, 2, 3, 6)] -> Bu, sort(v) ile aynı çıktıyı verir
````

Orijinal script'teki son satır (`sort(ind)`), `order` fonksiyonunun *çıktısını* (yani [5, 1, 4, 2, 3, 6] vektörünü) sıralar. Bu, genellikle `order`'ın kullanım amacı değildir ancak kodun ne yaptığını gösterir:

````r
sort(ind) # Yani [1, 2, 3, 4, 5, 6] sonucunu verir.
````
